```python
def preprocess(df):
    df = df.copy()
    
    def replace_fn(match):
        char = match.group(0)
        if char in ['.', "'", '-']:
            return ''
        elif char == 'š':
            return 's'
        return 'u'
    
    df = df[~df.name.str.contains(r"\w*\s+\w*")]
    df['name'] = df.name.str.lower()
    df['name'] = df.name.str.replace(r"['.\-šü]", replace_fn, regex=True)
    df['name'] = df.name.str.replace(r"(\w)\1", lambda match: match.group()[0], regex=True)
    df = df.drop_duplicates('name')

    return df

df = preprocess(df)
print(sorted(list(set(''.join(df.name)))))
```
